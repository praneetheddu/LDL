# Back Propagation using XOR Perceptron

```python
import numpy as np

np.random.seed(3)
LEARNING_RATE = 0.1


# Define training example

x_train = [np.array([1.0, -1.0, -1.0]) , np.array([1.0, -1.0, 1.0]), 
            np.array([1.0, 1.0, -1.0]), np.array([1.0, 1.0, 1.0])]
y_train = [0.0, 1.0, 1.0 , 0.0] # XOR output
index_list = np.arange(len(x_train))
```

```python
def sample_weights(n):
    weights = np.zeros(n + 1)
    for i in range(1, n+1):
        weights[i] = np.random.uniform(-1, 1)
    return weights

w_n = [sample_weights(2) , sample_weights(2) , sample_weights(2)]
y_hat = [0, 0, 0]
err = [0, 0, 0]
```

```python
def forward_pass(x):
    global y_hat
    y_hat[0] = np.tanh(np.dot(x, w_n[0])) # Neuron 1
    y_hat[1] = np.tanh(np.dot(x, w_n[1])) # Neuron 2
    n2_in = np.array([1.0 , y_hat[0], y_hat[1]]) 
    z = np.dot(n2_in, w_n[2])
    y_hat[2] = (1.0) / (1.0 + np.exp(-z))
    
    
def back_pass(y_truth):
    global err
    # sigmoid log and err derivative
    err_prime = -(y_truth - y_hat[2])
    der = y_hat[2] * (1.0 - y_hat[2])
    err[2] = err_prime * der
    
    # tanh derivative and err[2] 
    der = 1.0 - y_hat[0] ** 2
    err[0] = der * err[2] * w_n[2][1]
    der = 1.0 - y_hat[1] ** 2
    err[1] = der * err[2] * w_n[2][2]
    
def adjust_weights(x):
    global w_n
    w_n[0] += (x * LEARNING_RATE * err[0])
    w_n[1] += (x * LEARNING_RATE * err[1])
    n2_inputs = np.array([1.0 , y_hat[0], y_hat[1]]) # first term is bias
    w_n[2] += (n2_inputs * LEARNING_RATE * err[2])
```
